# Necessry Library --

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split,KFold,cross_val_score
from sklearn.linear_model import LinearRegression,Ridge ,Lasso
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error,r2_score
from sklearn.preprocessing import MinMaxScaler
```

# Data Import --

```python
df = pd.read_csv('/content/Clean_Insurance_Data')
```

# Splitting Data --

```python
df.head(2)
```

```python
columns = ['age', 'sex', 'bmi', 'children', 'smoker', 'BMI_Normal',
       'BMI_Under_Weight', 'BMI_obase', 'BMI_over_weight']
```

```python
scaler = MinMaxScaler()
df[columns] = scaler.fit_transform(df[columns])
```

```python
df.head(2)
```

```python
x = df.drop(columns = 'charges')
y =df['charges']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y , test_size=0.3,random_state=42)
```

```python
x_train, y_train =  x_train.values,  y_train.values
```

# Model Evaluation --

# Model Report

# Testing Report

# Training Report --

```python
poly = PolynomialFeatures(degree=5, include_bias=False)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)

# Lasso regression
lasso = Lasso(alpha=2, max_iter=50000)
lasso.fit(x_train_poly, y_train)

# Predictions
train_pred = lasso.predict(x_train_poly)
test_pred = lasso.predict(x_test_poly)

# Train metrics
train_r2 = r2_score(y_train, train_pred)
train_mae = mean_absolute_error(y_train, train_pred)
train_mse = mean_squared_error(y_train, train_pred)

# Test metrics
test_r2 = r2_score(y_test, test_pred)
test_mae = mean_absolute_error(y_test, test_pred)
test_mse = mean_squared_error(y_test, test_pred)

print("Train R2:", train_r2)
print("Test R2:", test_r2, "\n")

print("Train MAE:", train_mae)
print("Test MAE:", test_mae, "\n")

print("Train MSE:", train_mse, "\n")
print("Test MSE:", test_mse)
```
